```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn")
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
df.head()
```

```python
print(f"O número de linhas é {df.shape[0]}")
print(f"O número de colunas é {df.shape[1]}")
```

```python
df.info()
```

```python
print("Receita Total = {}".format(df["Valor Venda"].sum()))
```

```python
#Criando a coluna custo total
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head(5)
```

```python
print("Custo total = {}".format(round(df["custo"].sum(), 2)))
```

```python
#Criando a coluna lucro
df["lucro"]  = df["Valor Venda"] - df["custo"] 
df.head(3)
```

```python
print("Lucro total = {}".format(round(df["lucro"].sum(), 2)))
```

```python
#Coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]

df["Tempo_envio"] = df["Tempo_envio"].dt.days
df.head(3)
```

```python
print("Tempo médio de envia por Marca: {}".format(df.groupby("Marca")["Tempo_envio"].mean()))
```

```python
#verificando valores faltantes
df.isnull().sum()
```

```python
#Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
print("Total de produtos vendidos")
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
fig = plt.figure(figsize=(10,6))
produtos_vendidos = df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh()
plt.title("Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
plt.show()
```

```python
fig = plt.figure(figsize=(9,6))
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(color="yellowgreen")
plt.title("Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
plt.xticks(rotation=45)
plt.show()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df.loc[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
fig = plt.figure(figsize=(10,6))
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(marker="o")
plt.title("Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
df_2009.groupby("Marca")["lucro"].sum().plot.bar(color="red")
plt.title("Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
plt.show()
```

```python
fig = plt.figure(figsize=(9,6))
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
plt.show()
```

```python
fig = plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.boxplot(df["Tempo_envio"])
plt.subplot(1,2,2)
sns.distplot(df["Tempo_envio"])
plt.show()
```
